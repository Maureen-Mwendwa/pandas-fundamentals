```python
# Create a copy of the DataFrame
new_df = countries_df.copy()

# Drop rows where 'Birthrate' or 'Deathrate' are NaN
new_df.dropna(subset=['Birthrate', 'Deathrate'], inplace=True)

# Replace commas with dots and convert 'Birthrate' to numeric
new_df['Birthrate'] = new_df['Birthrate'].replace({',': '.'}, regex=True)
new_df['Birthrate'] = pd.to_numeric(new_df['Birthrate'], errors='coerce')

# Replace commas with dots and convert 'Deathrate' to numeric
new_df['Deathrate'] = new_df['Deathrate'].replace({',': '.'}, regex=True)
new_df['Deathrate'] = pd.to_numeric(new_df['Deathrate'], errors='coerce')

# Display DataFrame info
new_df.info()
```

```python
new_df[['Birthrate', 'Deathrate']].sum().plot(kind='pie', autopct='%1.1f%%', title='Proportion of Total Birthrate and Deathrate')
```

```python
# Stacked Bar Plot: A stacked bar plot can show how different categories contribute to a total across multiple groups
new_df.groupby('Region')[['Birthrate', 'Deathrate']].sum().plot(kind='bar', stacked=True, title='Sector Contributions by Region')
```

```python
countries_df.plot(kind='scatter', x='GDP ($ per capita)', y='Population', s=countries_df['Area (sq. mi.)'] / 1000, alpha=0.5, title='GDP vs Population with Area as Bubble Size')
```

```python
#Scatter Plots: Relationships Between Continuous Variables
countries_df.plot(kind='scatter', x='Pop. Density (per sq. mi.)', y='Coastline (coast/area ratio)', title='Population Density vs Coastline Ratio')

```
